```python
 # 読み込むデータが格納されたディレクトリのパス，必要に応じて変更の必要あり
path = "C:/Users/work/gci/competition3/"

df = pd.read_csv(path + 'train.csv')
df_test = pd.read_csv(path + 'test.csv')

```

```python
df['_hideDuration'] = df['matchDuration']-(round(df['walkDistance'] / 6) + round(df['rideDistance'] / 277) + df['swimDistance'] / 1)
df_test['_hideDuration'] = df_test['matchDuration']-(round(df_test['walkDistance'] / 6) + round(df_test['rideDistance'] / 277) + df_test['swimDistance'] / 1)

df['_headshotKillRate'] = df['headshotKills'] / df['kills']
df_test['_headshotKillRate'] = df_test['headshotKills'] / df_test['kills']

df['_killStreakrate'] = df['killStreaks']/df['kills']
df_test['_killStreakrate'] = df_test['killStreaks']/df_test['kills']

pd.isnull(df) 
```

```python
df_not_null = df.fillna(0)
df_test_not_null = df_test.fillna(0)
```

```python
df_delete = df_not_null.drop(['Id','groupId','matchId','matchType','winPlacePerc','walkDistance'], axis=1)
df_test_delete= df_test_not_null.drop(['Id','groupId','matchId','matchType','walkDistance'], axis=1)
```

```python
pd.isnull(df_delete) 
```

```python
df_concat = pd.concat([df_delete, df['winPlacePerc']], axis=1)
df_concat
```

```python
X = df_concat.iloc[:,:26].values
y = df_concat.iloc[:,26:].values
X_test = df_test_delete.iloc[:,:].values
```

```python
print(X_test.shape)
print(X.shape)
print(y.shape)
```

```python
#テストデータ
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.5, random_state=42)
print(type(y_train))
```

```python
rfc = RandomForestRegressor(max_depth=10, min_samples_leaf=1, n_estimators=100, n_jobs=-1, random_state=42)
rfc.fit(X_train, y_train)
print('Train Score: {}'.format(round(rfc.score(X_train, y_train), 3)))
print(' Test Score: {}'.format(round(rfc.score(X_valid, y_valid), 3)))
test = rfc.predict(X_valid)
mean_absolute_error(y_valid,test)
```

```python
y_ans = rfc.predict(X_test)
```

```python
path = "C:/Users/work/gci/competition3/"
submission = pd.read_csv(path + 'submission.csv')
submission['winPlacePerc'] = y_ans
submission.to_csv('C:/Users/work/gci/competition3/submission.csv', index=False)
submission
```
